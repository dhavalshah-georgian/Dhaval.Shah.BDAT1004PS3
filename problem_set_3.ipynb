```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
USMarrageDivorce = pd.read_csv('us-marriages-divorces.csv')
USMarrageDivorce.info()
```

```python
USMarrageDivorce.head()
```

```python
years = USMarrageDivorce['Year']
marriages = USMarrageDivorce['Marriages_per_1000']
divorces = USMarrageDivorce['Divorces_per_1000']
USMarrageDivorce = plt.figure(figsize= (16,8))
USMarrageDivorce = plt.plot(years, marriages, label ='Marriages')
USMarrageDivorce = plt.plot(years, divorces, label = 'Divorces')
USMarrageDivorce = plt.title("Number of marriages and divorces per capita in the U.S. between 1867 and 2014", fontsize=16)
USMarrageDivorce = plt.xlabel("Years", fontsize=14)
USMarrageDivorce = plt.legend(fontsize = 12, loc = "upper left")
#USMarrageDivorce = plt.ylabel("Marriages", fontsize=14)
USMarrageDivorce = plt.grid(True)
USMarrageDivorce
```

```python
#Question - 7
```

```python
AssignVar = USMarrageDivorce.loc[USMarrageDivorce.Year.isin([1900,1950,2000]), ['Year', 'Marriages_per_1000', 'Divorces_per_1000']]
AssignVar
```

```python
AssignVar = AssignVar[AssignVar['Year'].apply(lambda x: x in [1900, 1950, 2000])]
years = AssignVar['Year']
marriages = AssignVar['Marriages_per_1000']
divorces = AssignVar['Divorces_per_1000']
AssignVar = plt.figure(figsize= (16,8))
AssignVar = plt.bar(years, marriages, label ='Marriages')
AssignVar = plt.bar(years, divorces, label = 'Divorces')
AssignVar = plt.title("Number of marriages and divorces per capita in the U.S. between 1990,1950 and 2000", fontsize=16)
AssignVar = plt.xlabel("Years", fontsize=14)
AssignVar = plt.legend(fontsize = 12, loc = "upper left")
AssignVar
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ScatterPlot = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
ScatterPlot
```

```python
years = ScatterPlot['Year']
arcade = ScatterPlot['Total Arcade Revenue (billions)']
computerAward = ScatterPlot['Computer Science Doctorates Awarded (US)']
ScatterPlot = plt.figure(figsize= (16,8))
ScatterPlot = plt.scatter(years, arcade, label ='Total Arcade Revenue (billions)')
ScatterPlot = plt.scatter(years, computerAward, label = 'Computer Science Doctorates Awarded (US)')
ScatterPlot = plt.title("Total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.", fontsize=16)
ScatterPlot = plt.xlabel("Years", fontsize=14)
ScatterPlot = plt.legend(fontsize = 12, loc = "upper left")
#USMarrageDivorce = plt.ylabel("Marriages", fontsize=14)
ScatterPlot = plt.grid(True)
ScatterPlot
```
